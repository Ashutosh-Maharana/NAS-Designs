```python
#chaning to categorical data types
cat_columns = ['d_CabinetConstruction','d_BaseStructure','d_WeatherCertification','d_airflowtype_aston','d_basefloorcoating','d_baseinsulmat','d_BaseStructureHeight','d_DesignCFMMode_byBrand','d_ElectricalWiringType']
joined_df[cat_columns] = joined_df[cat_columns].astype('category')
#'d_BaseStructureHeight' : can be converted to int
# d_DesignEnvironment : only indoor and outdoor
#converting to binary
joined_df = joined_df.rename(columns = {'d_DesignEnvironment':'Is_Indoor','eGasHeatExchanger':'Is_eGasHeatExchanger','eCesFan':'Is_eCesFan'})
joined_df['Is_Indoor'] = joined_df['Is_Indoor'].map({'eIndoor':True, 'No':False})
joined_df['Is_Indoor'] = joined_df['Is_Indoor'].astype('bool')
joined_df['Is_eGasHeatExchanger'] = joined_df['Is_eGasHeatExchanger'].map({1:True, 0:False})
joined_df['Is_eGasHeatExchanger'] = joined_df['Is_eGasHeatExchanger'].astype('bool')
joined_df['Is_eCesFan'] = joined_df['Is_eCesFan'].map({1:True, 0:False})
joined_df['Is_eCesFan'] = joined_df['Is_eCesFan'].astype('bool')
```

```python
strong_corr = correlation_matrix.unstack().reset_index()
strong_corr.columns = [ 'Variable_1','Variable_2', 'Correlation']
strong_corr['pair'] = strong_corr.apply( lambda row: tuple(sorted([row['Variable_1'],row['Variable_2']])),axis = 1)
strong_corr = strong_corr[(strong_corr['Variable_1'] != strong_corr['Variable_2']) & (strong_corr['Correlation'].abs() >= 0.6)].drop_duplicates(subset = 'pair').sort_values(by = 'Correlation',ascending = False)
```
