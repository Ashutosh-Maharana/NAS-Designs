```python
# getting job hours to the df
excel_file = "C:\\Users\\Ashutosh.Maharana\\OneDrive - Nortek, Inc\\Projects\\linier_reg\\TB_corr_airtunnel_component_details.csv"
df_components = pd.read_csv(excel_file)
df_components.shape
```

```python
df_components.head(20)
```

```python

grouped_df = df_components.groupby(['ProjectID', 'VERSION', 'DashNo', 'DesignGuid','Component_name']).size().reset_index(name='Count')
```

```python
grouped_df.shape
```

```python
grouped_df.head(40)
```

```python

# povoting to get component counts
# Replace missing values with 0
pivot_df = grouped_df.pivot_table(index=['ProjectID', 'VERSION', 'DashNo','DesignGuid'],columns='Component_name', values='Count',fill_value=0).reset_index()
pivot_df.shape
```

```python
pivot_df.head(20)
```

```python
merged_df = pd.merge(merged_df, pivot_df, how='inner', left_on=['ProjectID', 'Version', 'DashNo','DesignGuid'], right_on = ['ProjectID',	'VERSION',	'DashNo',	'DesignGuid'	])
merged_df.shape
```
